```python

# Installation des dépendances
!pip install transformers torch numpy pandas matplotlib

# Import des bibliothèques
from transformers import pipeline
import re

# Initialisation du modèle NLP
print("Chargement du modèle d'extraction d'entités...")
nlp = pipeline("token-classification", model="dslim/bert-base-NER", aggregation_strategy="simple")

# Fonction d'extraction des dimensions
def extract_parameters(description):
    # Analyse avec le modèle
    results = nlp(description)

    # Extraction des entités pertinentes
    parameters = {
        "surface": None,
        "dimensions": [],
        "materiaux": [],
        "etages": None
    }

    # Recherche de motifs spécifiques
    # Surface
    surface_match = re.search(r'(\d+)\s*m²', description)
    if surface_match:
        parameters["surface"] = int(surface_match.group(1))

    # Étages
    etages_match = re.search(r'(\d+)\s*étages?', description)
    if etages_match:
        parameters["etages"] = int(etages_match.group(1))

    # Dimensions
    dimensions_matches = re.findall(r'(\d+)\s*cm', description)
    parameters["dimensions"] = [int(d) for d in dimensions_matches]

    # Matériaux
    materiaux = []
    if "béton" in description.lower():
        materiaux.append("béton")
    if "acier" in description.lower():
        materiaux.append("acier")
    if "bois" in description.lower():
        materiaux.append("bois")
    parameters["materiaux"] = materiaux

    return parameters

# Test
description_test = "Maison individuelle de 120m² avec 2 étages, murs en béton de 20cm d'épaisseur"
resultats = extract_parameters(description_test)
print("Description analysée:", description_test)
print("Paramètres extraits:", resultats)
```
